```python
best_seller = []
with open("./data/active_users.txt") as bs:
  bs.readline()
  for line in bs:
    best_seller.append(line.strip())
```

```python
niche = []
with open("./data/inactive_users.txt") as ni:
  ni.readline()
  for line in ni:
    niche.append(line.strip())
```

```python
diverse = []
with open("./data/medium_users.txt") as di:
  di.readline()
  for line in di:
    diverse.append(line.strip())
```

```python
len(best_seller), len(diverse), len(niche)
```

```python
import numpy as np

def ndcgk(actual, predicted):
  idcg = 1.0
  dcg = 1.0 if predicted[0] in actual else 0.0
  for i, p in enumerate(predicted[1:]):
    if p in actual:
      dcg += (1.0 / np.log2(i+3))
  return dcg / idcg

print(ndcgk([1], [3,1]))
```

```python
import pandas as pd
import ast
import math

results = pd.read_csv("./data/final_results.csv")
ndcg_per_user = pd.DataFrame(columns=["user_id", "ndcg", "type"])
print(results.columns)
results_per_user = results.groupby('user_id')
for user_id, group in results_per_user:
  ndcg = 0
  for eachrow in group.itertuples(index=False):
    try:
      ndcg+=ndcgk([str(eachrow.business_id)], ast.literal_eval(eachrow.recommended_list))
    except:
      pass
  ndcg /= len(group)
  if str(user_id) in niche:
    user_type = "N"
  elif str(user_id) in best_seller:
    user_type = "BS"
  else:
    user_type = "D"
  ndcg_per_user = ndcg_per_user.append({"user_id": user_id, "ndcg": ndcg, "type": user_type}, ignore_index=True)

print(ndcg_per_user)
```

```python
crs_heat_map = np.zeros((3,3))
ndcg_per_user_grouped = ndcg_per_user.groupby("type")
bs_n = 0
bs_d = 0
for user_id in best_seller:
  ndcg = ndcg_per_user[ndcg_per_user["user_id"]==int(user_id)]["ndcg"].to_list()[0]
  diverse_group = ndcg_per_user_grouped.get_group("D")
  bs_d += len(diverse_group[diverse_group["ndcg"]<ndcg])/len(diverse_group)

  niche_group = ndcg_per_user_grouped.get_group("N")
  bs_n += len(niche_group[niche_group["ndcg"]<ndcg])/len(niche_group)
  
crs_heat_map[0][1] = bs_d/len(best_seller)
crs_heat_map[0][2] = bs_n/len(best_seller)

d_bs = 0
d_n = 0
for user_id in diverse:
  ndcg = ndcg_per_user[ndcg_per_user["user_id"]==int(user_id)]["ndcg"].to_list()[0]
  best_seller_group = ndcg_per_user_grouped.get_group("BS")
  d_bs += len(best_seller_group[best_seller_group["ndcg"]<ndcg])/len(best_seller_group)

  niche_group = ndcg_per_user_grouped.get_group("N")
  d_n += len(niche_group[niche_group["ndcg"]<ndcg])/len(niche_group)

crs_heat_map[1][0] = d_bs/len(diverse)
crs_heat_map[1][2] = d_n/len(diverse)

n_bs = 0
n_d = 0
for user_id in niche:
  ndcg = ndcg_per_user[ndcg_per_user["user_id"]==int(user_id)]["ndcg"].to_list()[0]
  best_seller_group = ndcg_per_user_grouped.get_group("BS")
  n_bs += len(best_seller_group[best_seller_group["ndcg"]<ndcg])/len(best_seller_group)

  diverse_group = ndcg_per_user_grouped.get_group("D")
  n_d += len(diverse_group[diverse_group["ndcg"]<ndcg])/len(diverse_group)

crs_heat_map[2][0] = n_bs/len(niche)
crs_heat_map[2][1] = n_d/len(niche)

crs_heat_map
```

```python
import pandas as pd
from collections import defaultdict
from tqdm import tqdm

import cornac
from cornac.models import FM, MF, VAECF
from cornac.datasets import movielens
from cornac.eval_methods import BaseMethod, RatioSplit
```

```python
train = pd.read_csv("./data/train_ratings_filter.csv")
train.head()
```

```python
eval_method = BaseMethod.from_splits(
    train_data=train.to_numpy(),
    test_data=train.to_numpy(),
    val_data=train.to_numpy(), 
    rating_threshold=1.0, 
    exclude_unknowns=False, 
    verbose=True, 
    seed=123)
```

```python
models = [
    # MF()
    VAECF(k=10, autoencoder_structure=[20], act_fn="tanh", likelihood="mult", n_epochs=1000, batch_size=100, learning_rate=0.001, beta=1.0, seed=123, use_gpu=True, verbose=True)
    # FM(k0=1, k1=1, k2=8, method="sgd", max_iter=100000, seed=42, name="sgd"),
    # FM(k0=1, k1=1, k2=8, method="sgda", max_iter=100, seed=42, name="sgda"),
    # FM(k0=1, k1=1, k2=8, method="als", max_iter=100, seed=42, name="als"),
    # FM(k0=1, k1=1, k2=8, method="mcmc", max_iter=100, seed=42, name="mcmc"),
]
```

```python
exp = cornac.Experiment(eval_method=eval_method, models=models, metrics=[cornac.metrics.RMSE()], user_based=False)
exp.run()
```

```python
# users to recommended items dict
users_to_pred = defaultdict(list)

# user_idxs to user_ids: mapping from cornac to dataset
user_ids = list(eval_method.train_set.user_ids)
# item_idxs to item_ids: mapping from cornac to dataset
item_ids = list(eval_method.train_set.item_ids)

for model in exp.models:
  print(f"> Model: {model.name}")
  for uidx in tqdm(range(eval_method.total_users)):
    ordered_rec_items = list(reversed(model.score(uidx).argsort()))
    for rec_item in ordered_rec_items:
      users_to_pred[user_ids[uidx]].append(item_ids[rec_item])
```
